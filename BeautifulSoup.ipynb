# Beautiful Soup

Beautiful Soup es una libreria que te permite extraer un archivo HTML y navegar a traves de el y sus diferentes etiquetas. 
A estas aplicaciones se les llama Parser o analizadores sintacticos.

```python
from bs4 import BeautifulSoup
```

### Crear un objeto BeautifulSoup

Para crear un objeto en BeautifulSoup necesitamos tener el documento Html en un formato legible, preferentemente un String. Para esto podemos copiar el codigo de la web o usar la libreria **request** que lo extrae directamente. EJ:

```python
import requests
URL = "http://scrap.coredumped.es" 
page = requests.get(URL).text
page
```

Una vez obtenido el html como string, podemos crear el objeto. Usamos el constructor BeautifulSoup (nombreString,'html.parser') colocando el string y eligiendo el parser de HTML

```python
soup = BeautifulSoup(page, 'html.parser')
soup
```

## Find

Podemos encontrar facilmente la información que buscamos si conocemos algunos datos sobre ellas, como la etiqueta en la que se encuentra, que atributos tienen, que texto contiene, etc. Para ello usamos los metodos .find() y .find_all()

- .find() devuelve la primera concidencia de nuestra busqueda
- .find_all() devuelve todas las coincidencias de la busqueda

Ambos aceptan los mismos argumentos

**find_all(name, attrs, recursive, string, limit, ...)** 

Donde:
- name: Nombre de la etiqueta EJ: "h1", "div", "span"
- attrs : Atributos de la etiqueta Ej: id="nombre", class_="clase"
- string : El texto que contiene, exacto o como regex

Para usar correctamente find leer: (https://www.crummy.com/software/BeautifulSoup/bs4/doc/#find-all)

```python
capitals = soup.find_all("span", class_="country-capital")
print(capitals)
```
