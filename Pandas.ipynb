# Pandas

La libreria Pandas nos permite crear una Dataframe (una representacion de una tabla) y manejarla con relativa facilidad.

```python
import pandas as pd
```

## Creación de una Dataframe
Para crear crear una Dataframe de Pandas necesitamos tener nuestros datos en uno de los formatos aceptados y en un archivo o URL accesible:
- Una matriz, donde cada array que la compone es una entrada de la tabla
- En un archivo CSV, donde las entradas estan separadas por comas (","). Excel suele usar este formato
- En un archivo de texto, donde las entradas estan separadas por saltos de linea ("\n")

### Con matriz
Usamos el constructor **DataFrame(matriz, columns = nombresColumna)**. Recibe como primer argumento la matriz con las entradas, y luego el nombre de las columnas

```python
drivers = [['Hamilton', 103, 'UK'], ['Schumacher', 91, 'Germany'],
           ['Raikkonen', 26, 'Finland'], ['Alonso', 32, 'Spain'],
           ['Vettel', 53, 'Germany'], ['Sainz', 2, 'Spain'],
           ['Hakkinen', 20, 'Finland'], ['Senna', 41, 'Brazil'],
           ['Verstappen', 54, 'Netherlands'], ['Fittipaldi', 14, 'Brazil'],     # Sí, Fittipaldi existe
           [None, None, None], ['Sainz', 2, 'Spain']]

column_names = ['Driver Name', 'Number of Victories', 'Country']


df = pd.DataFrame(drivers, columns = column_names)
df
```

### Con archivo CSV desde URL
El Url contiene un archivo CSV con las entradas separadas por comas. Se guardas el Url en una variable auxiliar y se usa el metodo **read_csv(rutaArchivo)**. La ruta del archivo CSV puede ser un Url o una ruta local

```python
url = "https://gist.githubusercontent.com/w-dan/b84bdbfbd86b610a89aa0aa57e6efb5e/raw/aeaffaa9d26c66678b697ecb9fd99fd0c8eaca52/papers.csv"
df = pd.read_csv(url)

df
```
